## Selection sort

```python
import random

def selection_sort(list_to_sort):
    
    for i in range(len(list_to_sort)):
        minimum = list_to_sort[i]
        index_min = i
        for j in range(i+1, len(list_to_sort)):
            if minimum > list_to_sort[j]:
                minimum = list_to_sort[j]
                index_min = j
        if index_min != i:
            temp = list_to_sort[i]
            list_to_sort[i] = minimum
            list_to_sort[index_min] = temp
        else: 
            pass

    return list_to_sort

# Verification result on 10 random lists
for _ in range(10):
    print(selection_sort([random.randint(1,100000) for _ in range(10)]))
```

## Insertion sort

```python
import random 

def insertion_sort(list_to_sort, n):
    list_to_sort.append(random.randint(1,100000))
    list_to_sort.append(random.randint(1,100000))
    if list_to_sort[1] < list_to_sort[0]:
        list_to_sort[0] , list_to_sort[1] = list_to_sort[1], list_to_sort[0]

    list_of_random_number = [random.randint(1,10000) for _ in range(n-2)]
    for new_element in list_of_random_number:
        inserted = False
        for j in range(len(list_to_sort)-1):
            if new_element <= list_to_sort[j]:
                list_to_sort.insert(j, new_element)
                inserted = True
                break
            elif new_element >= list_to_sort[j] and new_element <= list_to_sort[j+1]:
                list_to_sort.insert(j+1, new_element)
                inserted = True
                break
            else:
                pass
        if not inserted:
            list_to_sort.append(new_element) 

    return list_to_sort

# Verification result on 10 random lists
for _ in range(10):
    print(insertion_sort([], 10))
```
