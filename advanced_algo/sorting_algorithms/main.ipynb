## Selection sort

```python
import random

def selection_sort(list_to_sort):
    for i in range(len(list_to_sort) - 1):
        min_idx = i
        for j in range(i+1, len(list_to_sort)):
            if list_to_sort[j] < list_to_sort[min_idx]:
                min_idx = j
        list_to_sort[min_idx], list_to_sort[i] = list_to_sort[i], list_to_sort[min_idx]
    return list_to_sort
# Verification
for _ in range(10):
    print(selection_sort([random.randint(1,100000) for _ in range(10)]))
```

## Insertion sort

```python
import random 

def insertion_sort(list_to_sort):
    for i in range(1, len(list_to_sort)):
        temp = list_to_sort[i]
        j = i-1
        while temp<list_to_sort[j] and j>=0:
            list_to_sort[j+1] = list_to_sort[j]
            j -= 1
        list_to_sort[j+1] = temp
    return list_to_sort
# Verification
for _ in range(10):
    print(insertion_sort([random.randint(1,10000) for _ in range(10)]))
```

## Bubble sort

```python
import random

def bubble_sort(list_to_sort):
    for _ in range(len(list_to_sort)):
        for j in range(len(list_to_sort)-1):
            if list_to_sort[j] > list_to_sort[j+1]:
                list_to_sort[j], list_to_sort[j+1] = list_to_sort[j+1], list_to_sort[j]
    return list_to_sort
# Verification
for _ in range(10):
    print(bubble_sort([random.randint(1,10000) for _ in range(10)]))
```

# Quick sort

```python
import random

def partition(list_to_sort, list_bottom, list_top):
    pivot = list_to_sort[list_top]
    i = list_bottom - 1
    for j in range(list_bottom, list_top):
        if list_to_sort[j] <= pivot:
            i += 1
            list_to_sort[i], list_to_sort[j] = list_to_sort[j], list_to_sort[i]
    list_to_sort[list_top], list_to_sort[i+1] = list_to_sort[i+1], list_to_sort[list_top]
    return i + 1

def quick_sort(list_to_sort, list_bottom = 0, list_top = None):
    if list_top is None:
        list_top = len(list_to_sort) - 1

    if(list_bottom < list_top):
        pivot_index = partition(list_to_sort, list_bottom, list_top)
        quick_sort(list_to_sort, list_bottom, pivot_index -1)
        quick_sort(list_to_sort, pivot_index + 1, list_top)

    return list_to_sort
# Verfication
for _ in range(10):
    print(quick_sort([random.randint(1,10000) for _ in range(10)]))
```
