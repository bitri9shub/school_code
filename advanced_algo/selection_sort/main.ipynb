# Sorting algorithms

## Selection sort

```python
import random

def selection_sort(list_to_sort):
    minimum = list_to_sort[0]
    for i in range(len(list_to_sort)):
        minimum = list_to_sort[i]
        for j in range(i+1, len(list_to_sort)):
            if minimum > list_to_sort[j]:
                minimum = list_to_sort[j]
                index_min = j
        temp = list_to_sort[i]
        list_to_sort[i] = minimum
        list_to_sort[index_min] = temp
    return list_to_sort

# Verification result on 10 random lists
for _ in range(10):
    random_list = [random.randint(1,100000) for _ in range(100)]
    print(selection_sort(random_list))
```
